```python
pl_2021_2021_matchs['Date_maths'] = pd.to_datetime(pl_2021_2021_matchs['Date'])
```

```python
pl_2021_2021_matchs['Date_maths']=pl_2021_2021_matchs['Date_maths'].dt.month
```

```python
pl_2021_2021_matchs_City = pl_2021_2021_matchs[(pl_2021_2021_matchs['HomeTeam']== 'Man City') ]
```

```python
Man_city_Home = pl_2021_2021_matchs_City.groupby(['HomeTeam', 'Date_maths'])['FTHG'].sum().reset_index()
```

```python
pl_2021_2021_matchs_City = pl_2021_2021_matchs[(pl_2021_2021_matchs['AwayTeam']== 'Man City') ]
```

```python
Man_city_Away = pl_2021_2021_matchs_City.groupby(['AwayTeam', 'Date_maths'])['FTAG'].sum().reset_index()
```

```python
Man_city_All = Man_city_Home.merge(Man_city_Away ,how= 'outer')
```

```python
Man_city_All.fillna(0, inplace=True)
```

```python
Man_city_All['nb_toatl'] = Man_city_All['FTHG'] + Man_city_All['FTAG']
```
